```python
import numpy as np
import pandas as pd
import os
```

# 1. Tum ogrencilerin bilgilerini tek bir dataframe icinde su sutunlardan olusacak sekilde birlestiriniz;
    # Sinif ismi,Ogrenci Adi, Dogru Sayisi, Yanlis Sayisi, Bos Sayisi" .

```python
files = os.listdir('./')
filesxlxs = [pd.ExcelFile(i) for i in files  if i.endswith ('.xlsx')] # bulundugumuz klasordeki tum excellleri exel formatinda aliyoruz
filesxlxs_names = [i[:-5] for i in files if i[-5:] =='.xlsx']
classes = []
student_names =[]
true = []
false = []
empty = []
count = 0

for i in filesxlxs:                              #tum excellerde dongu   
    for j in range(len(i.sheet_names)):          # excel icinde sayfalarda dongu
        classes +=[str(filesxlxs_names[count])]
        data = pd.read_excel(i, i.sheet_names[j])
        student_names+= [data.columns[0]]
        if  isinstance(data.iloc[-1,1],int):   #zerrin yanlis yere girilmis ama asagidan itibaren alinca problem cozulduu         
            true +=[data.iloc[-3,1]]
            false +=[data.iloc[-2,1]]
            empty += [data.iloc[-1,1]]           
        else:                                #Selcuk verileri girilmemis none atiyoruz
            true += [None]
            false += [None]
            empty +=[None]
    count += 1

students = pd.DataFrame({"Class": classes, 
                         "Name": student_names,
                         "Trues" :true,
                         "Falses": false,
                         "Empties": empty}) 

students.index = students.index +1
students
```

# 2.Sinava giren ogrenci sayisini hesaplayiniz.

```python
len(students)
```

# 3. Siniflarin ayri ayri dogru yanlis bos ortalamalari:

```python
table = pd.pivot_table(students, index = 'Class', values = ["Trues",'Falses','Empties'] ,fill_value=0)  #We can also fill missing values using the fill_value parameter.
table 
```

# 4.Her sinifin ayri ayri en basarili ogrencisini bulunuz.

```python
students["Result"] =students["Trues"]-students["Falses"]/4
students
```

```python
for i in filesxlxs_names:
    print(students.loc[students['Class'] == i].sort_values(by="Result",ascending=False).head(1))
```

```python
table = pd.pivot_table(students, index = ['Class', "Name"], values = "Result" ,aggfunc={"Result":np.max})
table
```

# 5. Tum sinavin en basarili ilk 3 ogrencisini bulunuz.

```python
students.sort_values(by="Result",ascending=False).head(3)
```

# 6. Siniflarin standart sapmalarini hesaplayiniz. (Dogru sayilari uzerinden)

```python
table = pd.pivot_table(students, index = ['Class'], values = "Trues" ,aggfunc={"Trues":np.std})
table
```
